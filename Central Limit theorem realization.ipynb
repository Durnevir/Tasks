```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
# fig, ax = plt.subplots(1, 1)
```

```python
x1 = np.linspace(0.0, 15.0)
b = 2.36
# зададим нормально распределенную случайную величину
pareto_rv = sts.pareto(b)
pdf_pareto = pareto_rv.pdf(x1)
# print (pdf_pareto)
# сгенерируем 10 значений
sample1 = pareto_rv.rvs(size = 1000)
# print sample1
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.plot(x1, pareto_rv.pdf(x1), 'k-', lw=1, label='probability density function', alpha = 0.6)
plt.hist(sample1, bins = 50, normed = True,  histtype = 'stepfilled', color = '#039F03', label = 'sample')
plt.plot(x1, pareto_rv.cdf(x1), 'r-', lw=1, label='cumulative distribution function', alpha = 1.0)
plt.legend()

standard_deviation = pareto_rv.std()
print 'среднеквадратическое отклонение равно', standard_deviation
mean_of_pareto = pareto_rv.stats(moments = 'm')
variance_of_pareto = pareto_rv.stats(moments = 'v')
print 'мат ожидание равно ', mean_of_pareto, '; дисперсия ', variance_of_pareto
standard_deviation_of_pareto = math.sqrt(variance_of_pareto)
print 'среднеквадратическое отклонение равно ', standard_deviation_of_pareto

x0 = np.linspace(1.65, 1.84, 10)
norm_rv_pareto = sts.norm(loc = mean_of_pareto, scale = standard_deviation_of_pareto)
```

```python
sample_means_pareto4 = []
for i in range(0, 1000):
    ssample_pareto = sts.pareto.rvs(b, loc=0.0, scale=1, size=4, random_state=None)
    sample_mean_pareto = np.mean(ssample_pareto)
    sample_means_pareto4.append(sample_mean_pareto)

minimal_4 = min(sample_means_pareto4)
maximal_4 = max(sample_means_pareto4)

print "мин знач", minimal_4
print "макс знач", maximal_4

x04 = np.linspace(minimal_4, maximal_4)
norm_rv_pareto_4 = sts.norm(loc = np.mean(sample_means_pareto4), scale = variance_of_pareto/4)

print "сред", np.mean(sample_means_pareto4)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.plot(x04, norm_rv_pareto_4.pdf(x04),'r-', lw = 1, label = 'pdf')
plt.hist(sample_means_pareto4, bins = 20, normed = True, histtype = 'stepfilled', color = '#C09AD7', label = "sample of size 4")
plt.legend()
```
