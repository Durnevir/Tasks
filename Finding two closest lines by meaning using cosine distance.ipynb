```python
import numpy as np
import re
f = open('sentences.txt', 'r')

text1 = f.read()
pp = text1.lower()
pp = re.split(r'[^a-z]', pp)
pp = [subelt for subelt in pp if subelt != "" ] # исключение пустых элементов их массива слов
frequencies = [] #массив с номерами, соответствующими частотам, с которыми слова встречаются в тексте
for i in range(len(pp)):
    frequencies.append(pp.count(pp[i]))
d = {}
for (i, j) in zip(pp, frequencies): #словарь, где ключом является слово, а элементом - частота, с кот. оно встречается
    d[i] = j
words = d.keys()


f = open('sentences.txt', 'r')
lines = f.readlines()
lineslow = []
for i in lines:
    lineslow.append(i.lower())

for j in range(len(lineslow)):
    lines[j] = re.split(r'[^a-z]', lineslow[j]);




# print lines    
#while True:
   # try:
#        lines.remove('')
#    except:
#        break
Rows = len(lines)
Columns = len(d)
lines = [[ subelement for subelement in element if subelement != "" ] for element in lines] 
matrix = np.ndarray(shape=(Rows, Columns)) #объявляем матрицу. число строк - это количество предложений в искомом тексте, число столбцов - количество слов, встречающих в тексте

for row in range(len(lines)):
    for element in range(len(d)):
        matrix[row][element] = lines[row].count(words[element]) #каждый элемент матрицы - это кол-во раз, с которым слово встретилось в предложении
# print matrix[1]

import scipy.spatial.distance
CosDist = []

for vec in range(len(matrix)-1):
    CosDist.append(scipy.spatial.distance.cosine(matrix[0], matrix[vec])) #определяем косинусное расстояние от первой строки до всех остальных. 
    print CosDist[vec]
```
